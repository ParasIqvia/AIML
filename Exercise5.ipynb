```python
#Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Import the dataset from this address.Assign it to a variable called iris
iris = pd.read_csv('iris.data', header=None)
```

```python
iris
```

```python
#Create columns for the dataset
iris.rename(columns={0:'sepal_length (in cm)',1:'sepal_width (in cm)',2:'petal_length (in cm)',3:'petal_width (in cm)',4:'class'},inplace=True)
```

```python
#Is there any missing value in the dataframe?
iris.isnull().sum()
```

```python
#Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN
iris['petal_length (in cm)'].iloc[9:29]=np.nan
iris['petal_length (in cm)'].iloc[9:29]
```

```python
#Good, now lets substitute the NaN values to 1.0
iris['petal_length (in cm)'].iloc[9:29].replace(np.nan,1.0,inplace=True)
iris['petal_length (in cm)'].iloc[9:29]
```

```python
#Now let's delete the column class
iris.drop('class',axis=1,inplace=True)
iris
```

```python
#Set the first 3 rows as NaN
iris.iloc[0:3]=np.nan
iris
```

```python
#Delete the rows that have NaN
iris.dropna(inplace=True)
```

```python
#Reset the index so it begins with 0 again
iris.reset_index(inplace=True)
```
